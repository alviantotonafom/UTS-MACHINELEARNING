### Essay :

#### 1. Apa itu Machine Learning?

**Machine Learning** (Pembelajaran Mesin) adalah cabang dari kecerdasan buatan (Artificial Intelligence) yang berfokus pada pengembangan algoritma dan model yang memungkinkan komputer untuk belajar dan membuat keputusan berdasarkan data empiris. Proses ini melibatkan beberapa langkah kunci, mulai dari pengumpulan data hingga penerapan model untuk membuat prediksi atau keputusan.

#### 2. Contoh Penerapan Implementasi dalam Kehidupan Sehari-hari  :

##### 1. Hasil Pencarian Search Engine
   - **Mempelajari Perilaku Pengguna** : Kata kunci yang sering digunakan, halaman yang paling sering dikunjungi.
   - **Interaksi Pengguna** : Memahami bagaimana pengguna berinteraksi dengan hasil pencarian sebelumnya.
   - **Hasil yang Relevan** : Menyediakan hasil pencarian yang lebih sesuai dengan apa yang dicari oleh pengguna.

##### 2. Rekomendasi Produk Marketplace
   - **Mempelajari Perilaku Belanja** : Produk yang dilihat, dibeli, atau ditambahkan ke keranjang belanja.
   - **Saran Produk** : Memberikan rekomendasi produk yang mungkin menarik minat pengguna berdasarkan data yang dikumpulkan.

##### 3. Pengalaman Pengguna Sosial Media
   - **Analisis Interaksi Konten** : Like, share, comment.
   - **Penyesuaian Feed** : Menampilkan konten yang dianggap paling relevan untuk pengguna.
   - **Deteksi Konten Melanggar** : Mendeteksi dan menghapus konten yang melanggar kebijakan platform.

##### 4. Konten yang Ditampilkan Periklanan Digital
   - **Analisis Data Pengguna** : Perilaku browsing, demografi, dan minat.
   - **Target Iklan** : Menampilkan iklan yang sesuai dengan preferensi pengguna untuk meningkatkan klik dan konversi.

##### 5. Asisten Pribadi Virtual
   - **Pengenalan Suara** : Mempelajari pola suara dan bahasa pengguna.
   - **Respons Akurat** : Menjawab pertanyaan atau menjalankan perintah dengan lebih akurat.
   - **Saran Personal** : Memberikan saran yang lebih personal berdasarkan preferensi pengguna.

##### 6. Fraud Detection
   - **Analisis Pola Transaksi** : Menganalisis pola transaksi dan perilaku pengguna.
   - **Identifikasi Anomali** : Mendeteksi anomali yang mungkin menunjukkan penipuan.
   - **Peringatan dan Blokir** : Memberikan peringatan atau memblokir transaksi yang mencurigakan untuk mencegah kerugian.

##### 7. Diagnosa Medis
   - **Analisis Data Pasien** : Menganalisis riwayat medis, hasil lab, dan gambar medis.
   - **Deteksi Penyakit** : Membantu dokter mendeteksi penyakit atau kondisi medis dengan tingkat akurasi yang tinggi.
   - **Peningkatan Diagnosa** : Meningkatkan akurasi dan kecepatan diagnosa medis.

##### 8. Mobil Kendali Otomatis
   - **Analisis Data Sensor** : Menganalisis data dari sensor, kamera, dan radar.
   - **Pemahaman Lingkungan** : Memahami lingkungan sekitar dan mendeteksi objek.
   - **Keputusan Mengemudi** : Membuat keputusan mengemudi seperti mempercepat, mengerem, atau berbelok.
   - **Pembelajaran dari Pengalaman** : Belajar dari pengalaman untuk meningkatkan keselamatan dan efisiensi berkendara.

#### 3. Macam-macam Taksononomi dalam Penerapan Machine Learning  :

![Taksonomi](../bahan/taksonomi.png)

##### Supervised (Pembelajaran Terawasi)
###### Linear (Linear)
- **Linear Regression** : Metode statistik untuk memprediksi nilai kontinu berdasarkan hubungan linear antara variabel input dan output.
- **Logistic Regression** : Digunakan untuk klasifikasi biner; memodelkan probabilitas kejadian tertentu dengan menggunakan fungsi logistik.
- **Perceptron** : Algoritma sederhana untuk klasifikasi biner yang menghasilkan hiperplan untuk memisahkan data.

###### Non-Linear (Non-Linear)
- **Single (Tunggal)**
  - **Decision Tree** : Algoritma yang memodelkan keputusan dan konsekuensi di dalam struktur pohon.
  - **Rule Learning** : Metode yang mengekstrak aturan dari data untuk membuat prediksi atau keputusan.
  - **Naive Bayes** : Klasifikasi probabilistik yang didasarkan pada teorema Bayes dengan asumsi independensi antar fitur.
  - **K-Nearest Neighbours (K-NN)** : Algoritma yang mengklasifikasikan data baru berdasarkan kedekatan dengan data dalam set pelatihan.
  - **Multi-Layer Perceptron** : Jaringan saraf buatan dengan lebih dari satu lapisan tersembunyi.
  - **Support Vector Machine (SVM)** : Algoritma yang mencari hiperplan terbaik yang memisahkan data ke dalam kelas yang berbeda.
  
- **Combined (Kombinasi)**
  - **Bagging** : Teknik ansambel untuk meningkatkan stabilitas dan akurasi.
  - **Boosting** : Metode ansambel untuk meningkatkan kinerja prediktif.
  - **Random Forest** : Algoritma ansambel untuk meningkatkan akurasi dan mengurangi overfitting.

##### Unsupervised (Pembelajaran Tak Terawasi)
- **K-Means** : Algoritma klasterisasi untuk membagi data ke dalam K klaster berdasarkan kedekatan.
- **Expectation-Maximization (EM)** : Algoritma untuk menemukan parameter model probabilistik dengan data tidak lengkap.
- **Self Organizing Maps (SOM)** : Jaringan saraf untuk memetakan data ke dalam peta dua dimensi.

##### Reinforced (Pembelajaran Penguatan)
- **Real-Time Decision** : Metode yang memungkinkan sistem membuat keputusan secara real-time.
- **Robot Navigation** : Teknik yang digunakan untuk mengajar robot bergerak dan berinteraksi dengan lingkungannya.
- **Learning Task** : Proses di mana agen belajar menyelesaikan tugas tertentu melalui trial and error.
- **Game AI** : Penerapan pembelajaran penguatan dalam mengembangkan kecerdasan buatan untuk permainan.
- **Skill Acquisition** : Proses di mana agen belajar dan mengakuisisi keterampilan baru.

```python
import pandas as pd

df = pd.read_csv('dataset-soalsk1.csv')

df.columns = df.columns.str.strip()

df['Biaya'] = df['Biaya'].apply(lambda x: int(x.split('x')[0]) * int(x.split('x')[1]))

rata_rata_datang = df['Datang'].mean()
print("a) Rata-rata mahasiswa datang pada minggu ini :", rata_rata_datang)

biaya_tertinggi = df.loc[df['Biaya'].idxmax()]
print("b) Biaya tertinggi terjadi pada hari :", biaya_tertinggi['Hari'])

hari_biaya_lebih_110000 = df.loc[df['Biaya'] > 110000, 'Hari'].tolist()
hari_biaya_lebih_110000_str = ', '.join(hari_biaya_lebih_110000)
print("c) Hari apa biaya lebih dari 110000 :", hari_biaya_lebih_110000_str)

mahasiswa_paling_banyak_datang = df['Mahasiswa'].mode().tolist()
mahasiswa_paling_banyak_datang_str = ', '.join(mahasiswa_paling_banyak_datang)
print("d) Mahasiswa yang paling banyak datang ke kampus :", mahasiswa_paling_banyak_datang_str)

mahasiswa_hari_minggu = df.loc[df['Hari'] == 'Minggu', 'Mahasiswa'].tolist()
mahasiswa_hari_minggu_str = ', '.join(mahasiswa_hari_minggu)
print("e) Mahasiswa yang datang pada hari Minggu :", mahasiswa_hari_minggu_str)

biaya_tertinggi = df['Biaya'].max()
biaya_terendah = df['Biaya'].min()
print("f) Biaya tertinggi:", biaya_tertinggi, "& Biaya terendah :", biaya_terendah)

frekuensi_datang_tertinggi = df['Mahasiswa'].value_counts().max()
frekuensi_datang_terendah = df['Mahasiswa'].value_counts().min()
print("g) Frekuensi datang tertinggi :", frekuensi_datang_tertinggi, "& Frekuensi datang terendah :", frekuensi_datang_terendah)
```
